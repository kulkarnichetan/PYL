# Seaborn:
- Seaborn is a Python data visualization library based on matplotlib. 
- It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.head()
```

```python
df.tail()
```

## Scatter Plot

```python
sns.scatterplot(x='age',y='fare',data=df)
```

## hue: 'hue' is used to visualize the data of different categories in one plot

## palette: 'palette' is used to change the colour of the plot.

```python
sns.scatterplot(x='age',y='fare',data=df,hue='class')
```

```python
sns.scatterplot(x='age',y='fare',data=df,hue='class', palette=["red","blue","Green"])
```

## pclass: 1-'Upper class', 2-'Middle class', 3-'Lower class'

```python
sns.scatterplot(x='age',y='fare',data=df,hue='class', palette=["red","blue","Green"], size='pclass')
```

## Histogram or Distribution Plot

```python
sns.histplot(df['age'],bins=5)  # bins- dividing into no. of parts
```

## KDE:
- A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. 
- KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
sns.histplot(df['age'],bins=10, kde=True)
```

## Bar Plot

```python
sns.barplot(x=df['sex'],y=df['fare'],palette=['red','blue'])
```

## Box Plot : It is more IMP in ML & DL
- minima, maxima, 25 percentile, 50 percentile,75 percentile)
- Data points which are above max value and below min value are called "Outliers"

```python
sns.boxplot(x=df['pclass'],y=df['fare'], palette=['yellow','red'])
```

```python
sns.boxplot(x=df['pclass'],y=df['fare'], palette=['yellow','red'], hue=df['sex'])
```

## Violin Plot

```python
sns.violinplot(x=df['pclass'],y=df['fare'], palette=['blue','red'])
```

```python
sns.violinplot(x=df['pclass'],y=df['fare'], palette=['blue','red'], hue=df['sex'])
```

## Strip Plot

```python
sns.stripplot(x=df['pclass'],y=df['fare'], palette=['blue','red'], hue=df['sex'])
```

```python
sns.stripplot(x=df['pclass'],y=df['fare'], palette=['blue','red'], hue=df['sex'], dodge=True)
```

## Joint Plot

```python
sns.jointplot(x=df['age'],y=df['fare'], palette=['blue','red'], hue=df['sex'], kind='kde')
```

```python
sns.jointplot(x=df['age'],y=df['fare'], palette=['blue','red'], hue=df['sex'], kind='kde', shade=True)
```

```python
sns.jointplot(x=df['age'],y=df['fare'], palette=['blue','red'], hue=df['sex'], kind='kde', shade=True, cmaps='coolwarm')
```

```python
sns.jointplot(x=df['age'],y=df['fare'], palette=['blue','red'], kind='hex', cmap='coolwarm')
```

## Pair Plot 

```python
df.columns
```

```python
sns.pairplot(df[["pclass",'fare','age','sibsp']],hue='sibsp')
```
